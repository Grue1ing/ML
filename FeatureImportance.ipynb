```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.simplefilter('ignore')
```

```python
Location = r'D:\git-project\project1\titanic.csv'
index_col = 'PassengerId'
data = pd.read_csv(Location) 
chooseData = np.array(['Pclass', 'Fare', 'Age', 'Sex'])
```

```python
print(data.info())
```

```python
del data['Parch']
del data['Name']
del data['SibSp']
del data['Ticket']
del data['Cabin']
del data['Embarked']
del data['PassengerId']
data
```

```python
data = data.dropna()
```

```python
print(data.info())
```

```python
target = data['Survived'].values
features = data[ ['Pclass', 'Fare', 'Age', 'Sex'] ].as_matrix()
```

```python
for i in range(714):
    if features[i][3] == "female":
        features[i][3] = 0
    else:
        features[i][3] = 1
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(features, target)
```

```python
importances = clf.feature_importances_
print(importances)
```
